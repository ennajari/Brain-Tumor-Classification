```python
import os
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
```

```python
train_dir = '../data/Training'
test_dir = '../data/Testing'
categories = ['glioma', 'meningioma', 'notumor', 'pituitary']
```

```python
for category in categories:
    category_dir = os.path.join(train_dir, category)
    num_images = len(os.listdir(category_dir))
    print(f"Nombre d'images dans la catégorie des données d'entraînement'{category}': {num_images}")
```

```python
for category in categories:
    category_dir = os.path.join(test_dir, category)
    num_images = len(os.listdir(category_dir))
    print(f"Nombre d'images dans la catégorie de test '{category}': {num_images}")
```

```python
plt.figure(figsize=(10, 10))
for i, category in enumerate(categories):
    category_dir = os.path.join(train_dir, category)
    image_files = os.listdir(category_dir)
    for j in range(3):  # Afficher 3 images par catégorie
        img_path = os.path.join(category_dir, image_files[j])
        img = Image.open(img_path)
        plt.subplot(4, 3, i * 3 + j + 1)
        plt.imshow(img)
        plt.title(f"{category} - {image_files[j]}")
        plt.axis('off')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 10))
for i, category in enumerate(categories):
    category_dir = os.path.join(test_dir, category)
    image_files = os.listdir(category_dir)
    for j in range(3):  # Afficher 3 images par catégorie
        img_path = os.path.join(category_dir, image_files[j])
        img = Image.open(img_path)
        plt.subplot(4, 3, i * 3 + j + 1)
        plt.imshow(img)
        plt.title(f"{category} - {image_files[j]}")
        plt.axis('off')
plt.tight_layout()
plt.show()
```

```python
for category in categories:
    category_dir = os.path.join(train_dir, category)
    sizes = []
    
    for img_file in os.listdir(category_dir):
        img_path = os.path.join(category_dir, img_file)
        img = Image.open(img_path)
        sizes.append(img.size)  # Ajoute la taille de l'image à la liste
    
    # Affiche les tailles moyennes et les écarts
    print(f"Catégorie : {category}")
    print(f"Taille moyenne des images : {np.mean(sizes, axis=0)}")
    print(f"Écart-type des tailles des images : {np.std(sizes, axis=0)}")
```
